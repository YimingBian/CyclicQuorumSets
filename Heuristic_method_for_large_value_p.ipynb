Decopose p into two factors p_1 and p_2. 

Inputs from user:

1. p, p_1, p_2 where p=(p_1)*(p_2)

2. two corresponding interest sets base_p1 and base_p2

Outputs: 

A set of quorums with all-pairs property


```python
#Inputs
p = 112
p1 = 7
p2 = 16

base_p1 = [0,1,3]
base_p2 = [0,1,2,5,8]
```

```python
def CQS_generator(p:int, base_set:list):
  cqs_list = list()
  for i in range(p):
    tmplist = list()
    for ii in base_set:
      tmplist.append((ii+i)%p)
    cqs_list.append(tmplist)
  return cqs_list
```

```python
#generate whole CQS based on two interest sets
cqs_p1 = CQS_generator(p1,base_p1)
cqs_p2 = CQS_generator(p2,base_p2)

#print(cqs_p1)
#print(cqs_p2)

#Regroup layer1
cqs_exp_p1 = list()
for i in cqs_p1: 
  tmplist = list()
  for ii in i:  
    for iii in range(p2): 
      tmplist.append(ii*p2+iii)
  cqs_exp_p1.append(tmplist)
#print(cqs_exp_p1)

#generate subset maps
subset_maps = list()
for i in cqs_exp_p1:  
  tmplist1 = list() 
  for ii in range(p2):  
    tmplist2 = list()
    for iii in range(len(base_p1)): 
       tmplist2.append(i[ii*len(base_p1)+iii])
    tmplist1.append(tmplist2)
  subset_maps.append(tmplist1)
  
#for i in subset_maps:
#  print(i,'\n')

#form the final set of quorums
from functools import reduce
from operator import add
final_quorum = list()
for map in subset_maps:
  for i in cqs_p2:
    tmplist = list()
    for ii in i:
      tmplist.append(map[ii])
    tmplist=reduce(add,tmplist)
    final_quorum.append(tmplist)
```

```python
#This function checks if a group of sets has all-pairs property.
#Pass p (int) and the whole group (list [[],[],...[]]) 
#(rather than the interest set because the group of quorums determined by thie heuristic method is not CQS)
def checker(p:int,sss:list):
  count1 = 0
  count2 = 0
  for i in range(p):
    for j in range(i+1,p):
      count2 = count2+1
      for ss in sss:
        if i in ss and j in ss:
          count1 = count1+1
          break
    if count1 == count2:
      return True
    else:
      return False
```

```python
#This function checks if a group of CQS has all-pairs property or not.
#Pass p (int) and a base set (list []), not necessarily an interest set
def CQS_checker(p:int,base_set:list):
  sss = CQS_generator(p,base_set)
  return checker(p,sss)
```

```python
# test CQS_checker()
print(CQS_checker(7,[0,1,2]))  # False
print(CQS_checker(7,[0,1,3]))  # True
print(CQS_checker(7,[1,2,4]))  # True
```
